- **Noticed**: Even though input raw image is b/w, the shape of numpy array is 3d. 
- The shape of grayscale images are 2d
- Hence, to work on grayscale images, it is necessary to convert 3d raw images to 2d images.

```python
fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(16, 9))
# fig, axs = plt.subplots(1,3)
plt.subplots_adjust(wspace=0.5)

axs[0].imshow(raw_im[:,:,0],cmap='gray')
axs[1].imshow(raw_im[:,:,1],cmap='gray')
axs[2].imshow(raw_im[:,:,2],cmap='gray')
```

- No difference is found between these images. Hence any axis/dimension can be selected. 

```python
def convert_image_to_grayscale(src_dir,dst_dir,im_filename):
    src_filename = src_dir + im_filename
    dst_filename = dst_dir + im_filename
    im = plt.imread(src_filename)
    plt.imsave(fname=dst_filename,arr=im[:,:,0],cmap='gray')
    print(f"Successfully converted image: {im_filename}")
```

```python
src_dir = '../images/grayscale/raw/'
dst_dir = '../images/grayscale/'
os.listdir(src_dir)
```

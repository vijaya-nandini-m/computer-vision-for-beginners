```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
im = plt.imread('../images/color/flowers/richard-kidger-aeCtUySmqks.jpg')
hsv_im = cv2.cvtColor(im,cv2.COLOR_RGB2HSV)
```

```python
fig, axs = plt.subplots(nrows=1,ncols=2,figsize=(16,9))

axs[0].imshow(im)
axs[0].set_title('Original Image')

axs[1].imshow(hsv_im)
axs[1].set_title('HSV Image')
```

```python
lower_red = np.array([0,120,70])    
upper_red = np.array([10,255,255])
mask1 = cv2.inRange(hsv_im,lower_red,upper_red)
```

```python
plt.imshow(mask1)
```

```python
np.unique(mask1)
```

```python
lower_red = np.array([165,120,70])    
upper_red = np.array([180,255,255])
mask2 = cv2.inRange(hsv_im,lower_red,upper_red)
```

```python
plt.imshow(mask2)
```

```python
np.unique(mask2)
```

```python
plt.imshow(mask1 | mask2)
```

```python
mask = mask1 | mask2
res_img = cv2.bitwise_and(im,im,mask=mask)
plt.imshow(res_img)
```

```python
lower_red = np.array([0,120,70])    
upper_red = np.array([8,255,255])
mask1 = cv2.inRange(hsv_im,lower_red,upper_red)

lower_red = np.array([170,120,70])    
upper_red = np.array([180,255,255])
mask2 = cv2.inRange(hsv_im,lower_red,upper_red)

mask = mask1 | mask2
res_img = cv2.bitwise_and(im,im,mask=mask)
plt.imshow(res_img)
```

```python
im_new = plt.imread('../images/color/flowers/matt-shalvatis-MczlpZGoCYA.jpg')
plt.imshow(im_new)
```

```python
hsv_img = cv2.cvtColor(im_new,cv2.COLOR_RGB2HSV)
plt.imshow(hsv_img)
```

```python
lower_yellow = np.array([20, 100, 100], np.uint8)
upper_yellow = np.array([40, 255, 255], np.uint8)
```

```python
yellow_mask = cv2.inRange(hsv_img, lower_yellow, upper_yellow)
plt.imshow(yellow_mask)
```

```python
yellow_result = cv2.bitwise_and(im_new,im_new,mask=yellow_mask)
plt.imshow(yellow_result)
```

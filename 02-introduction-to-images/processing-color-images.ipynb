```python
R = cim[:,:,0]
G = cim[:,:,1]
B = cim[:,:,2] 
```

```python
fig, axs = plt.subplots(nrows=1,ncols=3,figsize=(16,9))
# plt.subplots_adjust(wspace=0.3)
axs[0].imshow(R)
axs[0].set_title('Red')
axs[0].set_xticks([])
axs[0].set_yticks([])

axs[1].imshow(G)
axs[1].set_title('Green')
axs[1].set_xticks([])
axs[1].set_yticks([])

axs[2].imshow(B)
axs[2].set_title('Blue')
axs[2].set_xticks([])
axs[2].set_yticks([])
```

```python
def plot_color_image_1(cim,img_filename):
    fig, axs = plt.subplots(nrows=2,ncols=3,figsize=(16,9))
    # plt.subplots_adjust(wspace=0.3)
    axs[0,0].imshow(cim)
    axs[0,0].set_xticks([])
    axs[0,0].set_yticks([])

    axs[0,1].imshow(cim)
    axs[0,1].set_xticks([])
    axs[0,1].set_yticks([])

    axs[0,2].imshow(cim)
    axs[0,2].set_xticks([])
    axs[0,2].set_yticks([])

    axs[1,0].imshow(cim[:,:,0])
    axs[1,0].set_title('Red')
    axs[1,0].set_xticks([])
    axs[1,0].set_yticks([])

    axs[1,1].imshow(cim[:,:,1])
    axs[1,1].set_title('Green')
    axs[1,1].set_xticks([])
    axs[1,1].set_yticks([])

    axs[1,2].imshow(cim[:,:,2])
    axs[1,2].set_title('Blue')
    axs[1,2].set_xticks([])
    axs[1,2].set_yticks([])

    fig.savefig(fname=img_filename,dpi=400,bbox_inches='tight')
```

```python
def plot_color_image_2(color_img,img_filename):
    plt.subplot(231)
    plt.imshow(color_img)
    plt.xticks([])
    plt.yticks([])

    plt.subplot(232)
    plt.imshow(color_img)
    plt.xticks([])
    plt.yticks([])

    plt.subplot(233)
    plt.imshow(color_img)
    plt.xticks([])
    plt.yticks([])

    plt.subplot(234)
    plt.imshow(color_img[:,:,0])
    plt.title('Red')
    plt.xticks([])
    plt.yticks([])

    plt.subplot(235)
    plt.imshow(color_img[:,:,1])    # ,cmap='gray'
    plt.title('Green')
    plt.xticks([])
    plt.yticks([])

    plt.subplot(236)
    plt.imshow(color_img[:,:,2])
    plt.title('Blue')
    plt.xticks([])
    plt.yticks([])

    # plt.show()
    plt.savefig(img_filename,dpi=400,bbox_inches='tight')
```

```python
output_image = '../images/output/gerhard-crous-G_mYUvUHh_U__1.jpg'
cim = plt.imread("../images/color/gerhard-crous-G_mYUvUHh_U.jpg")
plot_color_image_1(cim,output_image)
```

```python
output_image = '../images/output/gerhard-crous-G_mYUvUHh_U__2.jpg'
cim = plt.imread("../images/color/gerhard-crous-G_mYUvUHh_U.jpg")
plot_color_image_2(cim,output_image)
```

### For Matplotlib color images are represented as RGB

```python
output_image = '../images/output/zoltan-tasi--noc08YBXi4__1_matplot.jpg'
cim1 = plt.imread('../images/color/zoltan-tasi--noc08YBXi4.jpg')
plot_color_image_1(cim1,output_image)
```

### OpenCV by default represents color images as BGR
- **Hence, Reversing numpy array to convert BGR to RGB**

```python
output_image = '../images/output/zoltan-tasi--noc08YBXi4__1_opencv.jpg'
cim2 = cv2.imread('../images/color/zoltan-tasi--noc08YBXi4.jpg')
cim2 = cim2[:,:,::-1]  ### Converting BGR to RGB
plot_color_image_1(cim2,output_image)
```
